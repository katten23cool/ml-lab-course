```python
graph = {
    'A': ['B' , 'C'] ,
    'B': ['D' , 'E'] , 
    'C': ['F'] ,
    'D': [] ,
    'E': ['F'] ,
    'F': []
}
```

```python
#BFS Breadth-FIrst Search
from collections import deque

def bfs ( graph , start , goal ) :
    queue = deque ([[ start ]])
    visited = set ()

    while queue :
        path = queue.popleft ()
        node = path[-1]
    
        if node == goal :
            return path
    
        if node not in visited :
            visited.add(node)
            for neighbor in graph [ node ]:
                new_path = list ( path )
                new_path.append ( neighbor )
                queue.append ( new_path )
        
    return None

print (" BFS Path : " , bfs ( graph , 'A' , 'F') )
```

```python
# Depth-First Search
def dfs(graph, start, goal):
    stack = [[start]]
    visited = set()

    while stack:
        path = stack.pop()
        node = path[-1]
        print(stack)

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)
            print(node)

            for neighbor in reversed(graph[node]):
                
                new_path = list(path)
                new_path.append(neighbor)
                stack.append(new_path)

    return None

print("DFS Path:", dfs(graph, 'A', 'F'))
```

```python
#A* Search Algorithm
import heapq

# Graph with costs
graph_cost = {
'A': { 'B': 1 , 'C': 3} ,
'B': { 'D': 1 , 'E': 3} ,
'C': { 'F': 5} ,
'D': {} ,
'E': { 'F': 1} ,
'F': {}
}

heuristic = {
'A': 5 ,
'B': 3 ,
'C': 4 ,
'D': 2 ,
'E': 1 ,
'F': 0
}

def a_star(graph, heuristic, start, goal):
    queue = [(0, start, [start])]
    visited = set()

    while queue:
        (cost, node, path) = heapq.heappop(queue)

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)

            for neighbor, weight in graph[node].items():
                new_cost = cost + weight + heuristic[neighbor]
                heapq.heappush(queue, (new_cost, neighbor, path + [neighbor]))

    return None


print("A* Path:", a_star(graph_cost, heuristic, 'A', 'F'))
    
```

```python
import heapq

graph = {
    'S': {'A': 1, 'B': 4},
    'A': {'C': 2},
    'B': {'C': 1},
    'C': {'G': 3},
    'G': {}
}

heuristic = {
    'S': 5,
    'A': 4,
    'B': 2,
    'C': 1,
    'G': 0
}

def a_star(graph, heuristic, start, goal):
    queue = [(0, start, [start])]
    visited = set()

    while queue:
        cost, node, path = heapq.heappop(queue)

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)

            for neighbor, weight in graph[node].items():
                new_cost = cost + weight + heuristic[neighbor]
                heapq.heappush(queue, (new_cost, neighbor, path + [neighbor]))

    return None

result = a_star(graph, heuristic, 'S', 'G')
print("A* Path:", result)
```

```python
import random

def f(x):
    return -(x**2) + 10 * x

def hill_climbing(start, step_size=1, max_iterations=10):
    current = start
    current_value = f(current)

    for i in range(max_iterations):

        neighbors = [current - step_size, current + step_size]

        next_node = current
        next_value = current_value

        for n in neighbors:
            value = f(n)
            if value > next_value:
                next_node = n
                next_value = value

        if next_value <= current_value:
            break

        current = next_node
        current_value = next_value

    return current, current_value


start_point = random.randint(20, 30)
solution, value = hill_climbing(start_point)

print("Start:", start_point)
print("Best solution x:", solution)
print("Max value f(x):", value)
```
